```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('/content/resumes_development.csv')
```

```python
data
```

```python
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
data
```

```python
X = data.drop('Interview', axis = 1)
y = data['Interview']
```

```python
X
```

```python
y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, train_size = 0.7, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler =StandardScaler()
scaler.fit(X_train)
X_train = pd.DataFrame(scaler.transform(X_train), index = X_train.index, columns = X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), index = X_test.index, columns = X_test.columns)
```

```python
X_train
```

```python
X_test
```

```python
models = {
    'Logistic Regression' : LogisticRegression(),
    'Decision Tree' : DecisionTreeClassifier(),
    'Random Forest' : RandomForestClassifier()
}

for name, model in models.items():
  model.fit(X_train, Y_train)
  print(name + ' trained!')
```

```python
for name, model in models.items():
  score = model.score(X_test, Y_test)
  print(name + ' score is: {:.2f}%'.format(score * 100))
```

```python
n_components = 20
pca = PCA(n_components = n_components)

pca.fit(X_train)

X_train_pca = pd.DataFrame(pca.transform(X_train), index = X_train.index, columns = ['PC' + str(i) for i in range(1, n_components + 1)])
X_test_pca = pd.DataFrame(pca.transform(X_test), index = X_test.index, columns = ['PC' + str(i) for i in range(1, n_components + 1)])
```

```python
X_train_pca
```

```python
X_test_pca
```

```python
for name, model in models.items():
  model.fit(X_train_pca, Y_train)
  print(name + ' trained!')
```

```python
for name, model in models.items():
  score = model.score(X_test_pca, Y_test)
  print(name + ' score is: {:.2f}%'.format(score * 100))
```
